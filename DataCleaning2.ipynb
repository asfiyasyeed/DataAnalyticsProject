```python
import pandas as pd
import glob

# Specify the directory containing the CSV files
directory_path = '/content/drive/MyDrive/data cleaning/*.csv'

# List all CSV files in the specified directory
csv_files = glob.glob(directory_path)

# Read all CSV files into a list of DataFrames with a specified encoding
dataframes = []
for file in csv_files:
    try:
        df = pd.read_csv(file, encoding='ISO-8859-1')  # Change the encoding if needed
        dataframes.append(df)
    except Exception as e:
        print(f"Error reading {file}: {e}")

# Optionally, concatenate them into a single DataFrame
combined_df = pd.concat(dataframes, ignore_index=True)

# Display the combined DataFrame
print(combined_df.head())
```

```python
for i in combined_df.select_dtypes(include="number").columns:
  df[i]=impute.fit_transform(df[[i]])
```

```python
def wisker(col):
  q1,q3=np.percentile(col,[25,75])
  iqr=q3-q1
  lw=q1-1.5*iqr
  uw=q3+1.5*iqr
  return lw,uw
```

```python
wisker(df['likes'])
```

```python
for i in ['views', 'likes', 'dislikes', 'comment_count']:
  lw,uw=wisker(df[i])
  df[i]=np.where(df[i]<lw,lw,df[i])
  df[i]=np.where(df[i]>uw,uw,df[i])
```

```python
for i in ['views', 'likes', 'dislikes', 'comment_count']:
  sns.boxplot(df[i])
  plt.show()
```
