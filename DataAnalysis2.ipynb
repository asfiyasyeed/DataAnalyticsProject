```python
high_calorie_items = data.nlargest(10, 'Calories')
print("\nTop 10 high-calorie items:")
print(high_calorie_items[['Item', 'Calories']])
```

```python
numeric_cols = [
    'Calories',
    'Total Fat',
    'Saturated Fat',
    'Trans Fat',
    'Cholesterol',
    'Sodium',
    'Carbohydrates',
    'Dietary Fiber',
    'Sugars',
    'Protein',
    'Vitamin A (% Daily Value)',
    'Vitamin C (% Daily Value)',
    'Calcium (% Daily Value)',
    'Iron (% Daily Value)'
]

# Convert columns to numeric, coercing errors
data[numeric_cols] = data[numeric_cols].apply(pd.to_numeric, errors='coerce')

# Now calculate the correlation matrix
correlation_matrix = data[numeric_cols].corr()
print("\nCorrelation matrix:")
print(correlation_matrix)

# Visualizing the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Calories'], bins=30, kde=True)
plt.title('Distribution of Calories')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Total Fat', data=data)
plt.title('Box Plot of Total Fat')
plt.xlabel('Total Fat (g)')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
category_counts = data['Category'].value_counts()
sns.barplot(x=category_counts.index, y=category_counts.values)
plt.title('Count of Items by Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
healthy_items = data[(data['Sugars'] < 5) & (data['Saturated Fat'] < 2)]
print(f"\nNumber of healthy items (low sugar & saturated fat): {len(healthy_items)}")
```

```python
print("\nSummary of findings:")
print("High-Calorie Items:", high_calorie_items[['Item', 'Calories']])
print("Healthy Items Sample:", healthy_items[['Item', 'Calories', 'Sugars', 'Saturated Fat']].head())
```
